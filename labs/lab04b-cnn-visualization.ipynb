# Nástroje na vizualizáciu CNN

### [COVNET PLAYGROUND](https://convnetplayground.fastforwardlabs.com/#/):
* Webová stránka s ukážkami 11 predtrénovaných konvolučných neurónových sietí + ukážka praktického použitia CNN - *image similarity search*.
* Netradičná vizualizácia - nezobrazuje ako jednotlivé príznakové mapy spracúvajú konkrétny vstup - zobrazuje ideálny vstup ktorý produkuje maximálne aktivačné hodnoty pre jednotlivé príznakové mapy.

### [COVNETJS](https://cs.stanford.edu/people/karpathy/convnetjs/):
* Javascriptová knižnica pre hlboké učenie s viacerými ukážkami dostupnými online, 2 z nich zobrazujú tréning konvolučných neurónových sietí. 
* Online ukážky umožňujú priamo upravovať topológiu siete, nahrávať a ukladať sieť. Pre offline použitie sa dá knižnica nainštalovať cez `npm` alebo stiahnuť ako `js` súbor.

### [TENSORSPACE](https://tensorspace.org/):
* Javascriptový framework založený na `tensorflow`. Online verzia ponúka 3D vizualizáciu viacero netrénovaných modelov.
* Offline verzia má obmedzenú funkcionalitu - použitie vlastnej siete vyžaduje pre-procesný nástroj.

### [NODELINK VISUALIZER](https://www.cs.ryerson.ca/~aharley/vis/):
* Javascriptová knižnica funkciami a vizuálne podobná tensorspace.
* Online ukážka má 2D aj 3D možnosť vizualizácie. 2D vie byť v niektorých prípadoch prehľadnejšie ako 3D. Js súbory sú dostupné ako zip, avšak pri použití sa vyskytujú problémy so zobrazovaním modelu. 

### [QUIVER](https://github.com/keplr-io/quiver):
* Python knižnica pre `keras` modely. Samotná vizualizácia je veľmi prehľadná a interaktívna, ale knižnica má rozsiahle inštalačné problémy na väčšine počítačov.

### [PICASSO](https://github.com/merantix/picasso):
* Podobné ako quiver.

# Tensorspace Playground

**Tensorspace Playground** je set interaktivných 3D ukážok, ktoré boli navrhnuté za účelom prezentovať rôzne pred-vytvorené a pred-trénované konvolučné neurónové siete. Tieto ukážky zahŕňajú rôzne modely hlbokého učenia, napr. klasifikáciu objektov, detegovanie objektov, generovanie obrazu, atď. Všetky modely v Playground sú interaktívne. 

Tieto modely musia byť pred použitím stiahnuté. Doporučuje sa používať Playground s lepším internetovým pripojením, keďže stiahnutie a načítanie väčších modelov môže chvíľu trvať (do niekoľko minút). 

Tensorspace odporúča používať zariadenia s minimálnou šírkou obrazovky aspoň 750 pixelov.

### Domovská stránka: https://tensorspace.org/
Domovská stránka má 3 tlačidlá. **“Get Started”** vedie k návodom pre používanie stiahnutej knižnice, **“Github”** k zdrojovému kódu a **“Playground”** k interaktívnym modelom. 

### Priamy link: https://tensorspace.org/html/playground/index.html 

## ROZLOŽENIE STRÁNKY:
![rozloženie](https://i.imgur.com/OfeOFNK.png)

Rozloženie individuálnych ukážok je veľmi podobné. 
1. **Väčší priestor na ľavo:** hlavný priestor, kde môžeme vidieť a interagovať s modelom. 
2. **Menší priestor na pravo:** nastavenia a panel s informáciami - tu môžeme meniť vstup a vidieť predikcie 
3. **Ľavý dolný roh:** tlačidlo na ľahký prístup k ostatným modelom

## 1. MODEL SIETE:

Prvá vrstva siete (vstupná vrstva => obraz na vstupe) je umiestnená na dolnej strane okna. Každá ďalšia vrstva je teda nad tou predošlou. 
### Ovládanie:
* kliknutie ľavým tlačidlom myši + potiahnutie - zmena uhla kamery 
* kliknutie pravým tlačidlom myši + potiahnutie - zmena pozície kamery 
* rolovanie kolieskom myši - priblíženie/oddialenie
* podržaním kurzora nad vrstvami môžeme vidieť vzťahy medzi nimi a veľkosti príznakových máp
* kliknutie na vrstvu ju expanduje (na tenzory/príznakové mapy) - nie všetky vrstvy sa dajú expandovať
* kliknutie na “mínus” vedľa expandovanej vrstvy ju vráti do pôvodného stavu
Pri RGB obrázkoch sa vstupný obrázok tiež dá expandovať na individuálne farebné kanály.

<!-- ![model](https://i.imgur.com/EMcIXpO.png) -->
<img src=https://i.imgur.com/EMcIXpO.png width="700">

### Farby:
* žltá - konvolučné vrstvy
* tyrkysová - pooling
* zelená - dense vrstvy 
* modrá - padding
* sivá - vstupná a výstupná vrstva (a YOLO Grid pre model *YOLOv2-tiny*)
* fialová - reshape
* ružová - aktivácia
* červená - dekonvolúcia
* oranžová - depthwise spatial convolution

## 2. NASTAVENIA:
V nastaveniach môžeme meniť vstup do siete, vidieť predikcie a meniť niektoré nastavenia. 

### Zmena vstupu:
Pri väčšine modelov sa v nastaveniach nachádza tlačidlo **“Select”**, pomocou ktorého môžeme vybrať vstup z predpripravenej vzorky. *LeNet* má priestor, do ktorého môžeme kresliť. *Acgan* má panel číslic, z ktorých je potrebné vybrať jedno a následne kliknúť na tlačidlo **“Generate”** aby bol obrázok číslice vygenerovaný a odoslaný na input. 

### Iné ovládacie prvky:
**“Reset”** - prepne kameru do východzieho nastavenia
*LeNet Training Demo* disponuje aj tlačidlami “Start”, “Stop”, “Reset” a “Change Test Image”

Tensorspace disponuje 8 ukážkami, z čoho 7 ukazuje pred-trénovanú sieť a jej predikciu pre vybraný vstup. *LeNet Training Visualization* ukazuje proces trénovania LeNet siete a trénovacie metriky pre vybranú triedu.

# CovnetJS

ConvnetJS je javascriptová knižnica pre hlboké učenie navrhnutá na Stanforde. [Kód knižnice je voľne dostupný na GitHube pod MIT licenciou](https://github.com/karpathy/convnetjs). 
ConvnetJS podporuje viaceré funkcionality neurónových sietí, vrátane schopnosti vytvoriť a natrénovať konvolučné neurónové siete pre spracovanie obrazu. 

Funkcionalita knižnice sa dá otestovať priamo vo viacerých interaktívnych ukážkach dostupných [tu](https://cs.stanford.edu/people/karpathy/convnetjs/). 

Pre konvolučné neurónové siete slúžia dve ukážky - MNIST demo a CIFAR-10 demo.

Stránka knižnica má taktiež aj sekciu s oficiálnou dokumentáciou a sekciu niekoľko úvodných tutoriálov pre stiahnutú knižnicu pod “Getting started”.

## 1. ONLINE UKÁŽKY: 
Obe spomínané ukážky zobrazujú trénovací proces konvolučnej siete na danom datasete. Trénovanie začína s načítaním stránky a môže byť kedykoľvek pozastavené aby sme detailne preskúmali vizualizáciu pre daný vstup. 
 
**MNIST demo:** 
* **Dataset:** obrázky s rozmerom *28x28* v odtieňoch sivej zobrazujúce číslice od 0-9
<!-- ![dataset_mnist](https://i.imgur.com/xSkOLEt.png) -->
<img src=https://i.imgur.com/xSkOLEt.png width="700">

* **Presnosť:** dosahuje okolo 99% po pár minútach
* **Augmentácia dát:** obrázky trénovacej množiny sú náhodne orezané na formát *24x24*
* Predikcia používa na klasifikáciu 4 náhodne orezané verzie obrázku  

**CIFAR-10 demo:**
* **Dataset:**  *32x32* RGB obrázky z 10 tried zvierat a dopravných prostriedkov (loď, žaba, lietadlo, auto, pes…)
<!-- ![dataset_cifar](https://i.imgur.com/bWfEuQL.png) -->
<img src=https://i.imgur.com/bWfEuQL.png width="700">

* **Presnosť:** ku koncu trénovania okolo 90%
* **Augmentácia dát:** náhodné prevracanie a posúvanie obrázkov o max 2px horizontálne a vertikálne

Oba datasety sú parsenuté python kódmi dostupnými v popise danej ukážky. 
Obe ukážky používajú ako optimalizátor Adadelta pokiaľ nenastavíte inak.

### ROZLOŽENIE UKÁŽOK: 
Oba interaktívne ukážky majú rovnaké všeobecné rozloženie.
Zhora stránky nadol:
1. **Sekcia trénovacích údajov (Training stats):** hlavné metriky trénovania siete, niektoré nastaviteľné užívateľom, + graf.
2. **Sekcia vytvorenia siete a metódy trénovania (Instantiate a network and trainer):** textové pole kde viete priamo nakódovať vašu sieť.
3. **Sekcia vizualizície (Network visualization):** zobrazenie jednotlivých vrstiev a ich parametrov. 
4. **Ukážky predikcie (Example predictions):** predikované triedy pre spracované obrázky.

### 1. SEKCIA TRAINING STATS: 
<!-- ![training_stats](https://i.imgur.com/37FZmwQ.png) -->
<img src=https://i.imgur.com/37FZmwQ.png width="600">

#### Informatívne údaje:
* **forward pass time** a **backward pass time** - čas v ms pre daný vstup
* Stratová funkcia - **classification loss** (taktiež zobrazené v grafe)
* Stratová funkcia zoslabovania váh - **L2 weight decay loss**
* Presnosť siete pre trénovaciu a validačnú množinu - **training/validation accuracy**
* Počet spracovaných vstupov - **examples seen**
 
#### Nastavitelné parametre:
* Učiaci parameter - **learning rate**
* Momentum
* Batch size
* Zoslabovanie váh - **weight decay**
 
Tieto parametre sa dajú meniť kedykoľvek a prejavia sa od nasledujúceho spracovaného vstupu (v ukážke predikcií si môžete všimnúť pokles/nárast presnosti) 
 
Obe ukážky používajú automaticky Adadelta optimalizátor, ale ak chcete vyskúšať aj SDG, viete taktiež meniť momentum a  parameter zoslabovania váh. 
 
Používateľ vie taktiež:
* **Vyčistiť graf** - graf chybovej funkcie voči počtu spracovaných vstupov
* **Natiahnuť alebo uložiť snapshot siete ako JSON**
* **Natiahnuť pred-trénovanú sieť (iba v CIFAR-10 deme).**

### NATIAHNUTIE A ULOŽENIE SIETE: 
 
V paneli trénovacích údajov nájdete pre tento účel dve tlačítka a textové pole. Veľkosť textového poľa je meniteľná. 

#### Tlačítka:
* **Save network snapshot as JSON** poskytne momentálne parametre siete ako plain text v textovom okne, kde sa dá potom skopírovať a uložiť.
* **Init network from JSON snapshot** nainicializuje sieť podľa parametrov v textovom okne. Nesprávny vstup je ignorovaný. 
 
#### Formát vstupu: 
Aj defaultná jednoduchá sieť v MNIST ukážke vytvorí snapshot dlhý niekoľko strán, keďže tento snapshot zahŕňa špecifické nastavenia všetkých váh. 
Preto odporúčame používať túto funkcionalitu iba na uloženie natrénovanej siete z ukážky alebo načítanie JSONu uloženej siete vytvorenej touto ukážkou.   
 
**Na navrhnutie vlastnej topológie siete použite textové pole v paneli “Instantiate a network and trainer”.**

## 2. JAVASCRIPTOVÁ KNIŽNICA: 

Knižnica sa dá použiť v prehliadači, alebo na serveri cez `node.js`.

### INŠTALÁCIA:

#### NODE.JS:
Knižnica sa dá stiahnuť cez `npm`.

Návod:
1. **Inštalácia:** `$ npm install convnetjs`
2. **Použitie:** zahrňte `var convnetjs = require("convnetjs");` vo svojom kóde.

**Direct link na npm package:** https://www.npmjs.com/package/convnetjs

#### V PREHLIADAČI: 
Minifikovaná knižnica je dostupná cez tento <a href=https://cs.stanford.edu/people/karpathy/convnetjs/build/convnet-min.js>link</a>, alebo kompletný kód cez <a href=https://github.com/karpathy/convnetjs/releases>GitHub</a> (minifikovaná aj neminifikovaná knižnica). 

Návod:

2. Zahrňte odkaz na váš .js script alebo ho napíšete do tela dokumentu. 

### FUNKCIE KNIŽNICE - PREHĽAD:

Knižnica má svoju oficiálnu dokumentáciu na https://cs.stanford.edu/people/karpathy/convnetjs/docs.html ako aj poznámky v samotnom kóde. Toto je všeobecný prehľad funkcionality ktorú môžete použiť v rámci ukážok alebo vlastného kódu.

Knižnica má 4 hlavné, dôležité triedy - **triedu vols, triedu vrstiev, triedu siete,** a **triedu trénovacích metód.**
Knižnica má zopár ďalších dodatočných tried s užitočnými funkciami pre vizualizáciu a podobne. 

#### Vol Class:
Hlavný stavebný kameň knižnice. 3D volume čísel o šírke, výške a hĺbke. Drží dáta pre všetky filtre, gradienty, váhy…
Pri písaní vlastnej topológie v rámci ukážok ju priamo nepoužijete.

#### Net class:
Funkcionalita neurónovej siete. V podstate zoznam vrstiev a funkcie nad týmto zoznamom.  

**Hlavné funkcie:**
* **makelayers(layer_defs)** - prijíma definíciu vrstiev (zoznam), vytvorí sieť o týchto vrstvách
* **getCostLoss(V,y)** - vykoná dopredný a spätný chod, vráti hodnotu chybovej funkcie
* **getPrediction()** - vráti predikovanú triedu 
* **getParamsAndGrads()** - naakumuluje a vráti parametre a gradienty z celej siete (pre vizualizačné účely)
* **toJSON()** - uloží sieť ako JSON
* **romJSON(json)** - natiahne sieť z JSONu

Pri používaní vlastnej siete v demách treba len zadeklarovať sieť a použiť funkciu makelayers(). Všetko ostatné sa vykonáva ako súčasť funkcionality ukážky. 

#### Layers class:
Knižnica podporuje vrstvy vstupné, fully connected, loss (klasifikačná a regresná verzia), konvolučné, pooling vrstvy a local contrast normalization vrstvy.

Na vytvorenie vrstiev a ich pridanie do siete vytvorte prázdny zoznam `(layer_defs_list = [])` a pridajte vrstvy funkciou `layer_defs_list.push({-layer parameters-})`. 

*Vzorový kód (pridáva konvolučnú vrstvu):*
`layer_defs.push({type:'conv', sx:5, filters:8, stride:1, pad:2, activation:'relu'});`

#### Trainers class:
Knižnica podporuje optimalizátory Adeldata, Adagrad a SDG, špecifikované pomocou parametra “method”. 

*Vzorový kód (inicializuje optimalizátor typu SDG):*
`var trainer = new convnetjs.Trainer(net, {method: 'sgd', learning_rate: 0.01,l2_decay: 0.001, momentum: 0.9, batch_size: 10, l1_decay: 0.001});`

Nepovinné parametre zahŕňajú *learning rate, l1 a l2 decay, batch size, momentum, ro (adeltada), eps (adeldata & adam), beta1 a beta2 (adam)*. 

Hlavné funkcie:
* train(x,y). 

V rámci ukážok stačí trainer zadeklarovať, tréning samotný už vykonáva demo (t.j. Funkciu train v ukážkach nepoužijete).
